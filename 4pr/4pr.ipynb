```python
import pandas as pd

read = pd.read_csv('insurance.csv')

data = pd.DataFrame(read)

data
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

data.hist(color='green', edgecolor='black')
plt.show()
```

```python
import numpy as np
import statistics as sts
import scipy.stats as ss

mean_bmi = np.mean(data['bmi'])
moda_bmi = sts.mode(data['bmi'])
med_bmi = np.median(data['bmi'])


print('Среднее bmi = ' f'{mean_bmi}')
print('Мода bmi = ' f'{moda_bmi}')
print('Медиана bmi = ' f'{med_bmi}')

fig,ax=plt.subplots()

ax.bar('Среднее', mean_bmi, label = 'Среднее')
ax.bar('Мода', moda_bmi, label = 'Мода')
ax.bar('Медиана', med_bmi, label = 'Медиана')

ax.legend()
plt.show()
```

```python
mean_ch = np.mean(data['charges'])
moda_ch = sts.mode(data['charges'])
med_ch = np.median(data['charges'])


print('Среднее charges = ' f'{mean_ch}')
print('Мода charges = ' f'{moda_ch}')
print('Медиана bmi = ' f'{med_ch}')

fig,ax=plt.subplots()

ax.bar('Среднее', mean_ch, label = 'Среднее')
ax.bar('Мода', moda_ch, label = 'Мода')
ax.bar('Медиана', med_ch, label = 'Медиана')

ax.legend()
plt.show()
```

```python
std_bmi = data['bmi'].std()
raz_bmi = data['bmi'].max() - data['bmi'].min()
q1_bmi = np.percentile(data['bmi'], 25, interpolation='midpoint')
q3_bmi = np.percentile(data['bmi'], 75, interpolation='midpoint')
iqr1_bmi = q3_bmi - q1_bmi
iqr2_bmi = ss.iqr(data['bmi'], interpolation='midpoint')


print('\nСтандартное отклонение: ', std_bmi)
print('Размах: ', raz_bmi)
print('Межквартильный размах через numpy: ', iqr1_bmi)
print('Межквартильный размах через scipy: ', iqr2_bmi)

fig, ax = plt.subplots()
ax.bar('Стандартное отклонение bmi', std_bmi, label='Стандартное отклонение bmi')
ax.bar('Размах bmi', raz_bmi, label='Размах bmi')
ax.bar('Межквартильный размах bmi', iqr2_bmi, label='Межквартильный размах bmi')
ax.legend()
plt.show()
```

```python
std_charges = data['charges'].std()
raz_charges = data['charges'].max() - data['charges'].min()
q1_charges = np.percentile(data['charges'], 25, interpolation='midpoint')
q3_charges = np.percentile(data['charges'], 75, interpolation='midpoint')
iqr1_charges = q3_charges - q1_charges
iqr2_charges = ss.iqr(data['charges'], interpolation='midpoint')

print('\nСтандартное отклонение: ', std_charges)
print('Размах: ', raz_charges)
print('Межквартильный размах через numpy: ', iqr1_charges)
print('Межквартильный размах через scipy: ', iqr2_charges)

fig, ax = plt.subplots()
ax.bar('Стандартное отклонение charges', std_charges, label='Стандартное отклонение charges')
ax.bar('Размах charges', raz_charges, label='Размах charges')
ax.bar('Межквартильный размах charges', iqr2_charges, label='Межквартильный размах charges')
ax.legend()
plt.show()
```

```python
a = np.array(data['bmi'])
sample_mean=[]
sample_std=[]
samples=[]
for i in range(300):
    sample=[]
    for j in range(150):
        sample.append(a[int(np.random.random()*len(a))])
    sample = np.array(sample)
    sample_mean.append(sample.mean())
    sample_std.append(sample.std())
    samples.append(sample)
sample_mean_np = np.array(sample_mean)
sample_std_np = np.array(sample_std)
plt.figure(figsize=(20,10),dpi=80)
d =0.1
num_bins = int((max(sample_mean_np)-min(sample_mean_np))//d)
print(num_bins)
print(sample_mean_np.mean())
print(sample_mean_np.std())
plt.hist(sample_mean_np, num_bins)
plt.show()
```
